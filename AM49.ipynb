```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_zomato=pd.read_csv(r"C:\Users\HP\Downloads\zomato.csv",encoding='latin-1')
df_zomato
```

```python
df_countryCode=pd.read_excel(r"C:\Dataset\Country-Code.xlsx")
df_countryCode
```

```python
df_final=pd.merge(df_zomato,df_countryCode,on='Country Code',how='left')
df_final
```

```python
df_final.info()
```

```python
counts=df_final.Country.value_counts()
counts
```

```python
country_name=df_final.Country.value_counts().index
country_name
```

```python
plt.bar(country_name[:3],counts[:3])
plt.xlabel("Country Name")
plt.ylabel("Counts")
plt.title("Top 3 Countries that uses Zomato Services")
plt.show()
```

```python
plt.plot(country_name[:3],counts[:3])
plt.xlabel("Country Name")
plt.ylabel("Counts")
plt.title("Top 3 Countries that uses Zomato Services")
plt.show()
```
